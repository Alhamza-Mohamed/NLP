```python
def basic_hash_table(value_l, n_buckets):
#value_l: a list of numbers
#n_buckets: how many buckets the hash table should have

    def hash_function(value, n_buckets):
        return int (value) %n_buckets # tells which bucket a value should go into
    """
    Example:
    value = 10, buckets = 5 → 10 % 5 = 0  
    value = 12, buckets = 5 → 12 % 5 = 2  
    value = 17, buckets = 5 → 17 % 5 = 2  

    """
    hash_table = {i:[] for i in range (n_buckets)} # Intialize all the buckets in the hash table as empty lists

    for value in value_l:
        hash_value = hash_function(value, n_buckets) # gives the bucket index
        hash_table[hash_value].append(value)  # Add the value to the corresponding bucket

    return hash_table
```

```python
value_l = [100, 10, 14, 17, 97] # Set of values to hash
hash_table_example = basic_hash_table(value_l , n_buckets=10)
pp.pprint(hash_table_example)
```

```python
def side_of_plane(P,v):
    dotproduct = np.dot(P,v.T) # Get the dot product
    sign_of_dot_product = np.sign(dotproduct) #the sign of the elements of the dotproduct matrix
    sign_of_dot_product_scalar = sign_of_dot_product.item() # The value of the first item
    return sign_of_dot_product_scalar
```

## Random planes

```python
np.random.seed(0)
num_dimensions = 2 
num_planes = 3 
random_planes_matrix = np.random.normal(size = (num_planes, num_dimensions))
print (random_planes_matrix)
```

```python
v = np.array([[2,2]])
```

```python
# side of the plane function. the result is matrix
def side_of_plane_matrix (P,v):
    dotproduct = np.dot(P,v.T)
    sine_of_dot_product = np.sign(dotproduct) # Get a boolean value telling if the value in the cell is positive or negative
    return sine_of_dot_product
```

```python
sides_l = side_of_plane_matrix(random_planes_matrix, v)
sides_l
```

```python
def hash_multi_plane_matrix(P,v,num_planes):
    sides_matrix = side_of_plane_matrix(P,v) # Get the side of planes for p and v
    hash_value = 0
    for i in range (num_planes):
        sign = sides_matrix[i].item() # Get the value inside the matrix cell
        hash_i = 1 if sign >= 0 else 0
        hash_value += 2**i * hash_i # sum 2^i * hash_i

    return hash_value
```

```python
hash_multi_plane_matrix(random_planes_matrix, v ,num_planes)
```
